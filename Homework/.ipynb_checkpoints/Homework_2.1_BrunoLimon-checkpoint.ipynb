```python
# libraries needed for haversine function
from math import sin, cos, sqrt, atan2, radians
# for plotting
import matplotlib.pyplot as plt
```

```python
# Approx radius of earth (km)
r = 6371

# Latitutes and Longitudes of Pisa and 5 largest capitals of Europe (according to population)
# Converted to radians since python trigonometric functions use radians by default
latPisa = radians(43.7194008)
lonPisa = radians(10.4142611)
latLondon = radians(52.406374)
lonLondon = radians(16.9251681)
latBerlin = radians(52.5170365)
lonBerlin = radians(13.3888599)
latMadrid = radians(40.4167047)
lonMadrid = radians(-3.7035825)
latRome = radians(41.8933203)
lonRome = radians(12.4829321)
latParis = radians(48.856614)
lonParis = radians(2.352221)

# List of Cities
cities = ["Pisa",
          "London",
          "Berlin",
          "Madrid",
          "Rome",
          "Paris"]  

# List of latitudes
latitudes = [latPisa,
             latLondon,
             latBerlin,
             latMadrid,
             latRome, 
             latParis]

# List of longitudes
longitudes = [lonPisa,
              lonLondon,
              lonBerlin,
              lonMadrid,
              lonRome,  
              lonParis]
```

```python
earth_distances = [] # empty list for earth distances
# For loop to compute the haversine function on each capital against Pisa
for i in range(1,6):
    haversine = r * (2 * atan2(sqrt(sin((latitudes[0] - latitudes[i]) / 2)**2 + cos(latitudes[0]) * cos(latitudes[i]) * sin((longitudes[0] - longitudes[i]) / 2)**2), 
                               sqrt(1 - sin((latitudes[0] - latitudes[i]) / 2)**2 + cos(latitudes[0]) * cos(latitudes[i]) * sin((longitudes[0] - longitudes[i]) / 2)**2)))
    earth_distances.append(haversine)

for i in range(0,5):
    print("The earth distance between", cities[0], "and", cities[i+1], "=", round(earth_distances[i],2), "km")
```

```python
eucl_distances = []  # empty list for euclidean distances
# For loop to compute the euclidean function on each capital against Pisa
for i in range(1,len(cities)):
    eucl = r * sqrt((latitudes[0] - latitudes[i])**2 + (longitudes[0] - longitudes[i])**2) 
    eucl_distances.append(eucl)
    
for i in range(0,5):
    print("The euclidean distance between", cities[0], "and", cities[i+1], "=", round(eucl_distances[i],2), "km")
```

```python
differences = []
# Absolute difference in values of earth and euclidean distances
for i in range(0,len(earth_distances)):
    dif = abs(earth_distances[i] - eucl_distances[i])
    differences.append(dif)
differences
```

```python
# Bar chart with the absolute difference between the Earth and Euclidean distances for each capital
plt.bar(["London","Berlin","Madrid","Rome","Paris"], differences)
plt.title("Difference between earth and euclidean distance")
plt.ylabel("km")
plt.show()
```

```python
# Comparison between the Earth and Euclidean distances for each capital
# Plot useful to see if different distant functions are coherent between each other and also see the difference in results as
# in the last bar chart, here we can confirm the share the same trend, even though with the madrid distance having a considerable gap
plt.plot(["London","Berlin","Madrid","Rome","Paris"], earth_distances, '.-', markersize=10, label="earth")
plt.plot(range(0,5), eucl_distances, '.-', label="euclidean", markersize=10)
plt.title("Earth and euclidean distances")
plt.ylabel("km")
plt.legend(loc='best')
plt.show()
```
